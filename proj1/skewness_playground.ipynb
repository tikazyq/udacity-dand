```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
import statsmodels as sm
from statsmodels.graphics.gofplots import qqplot, qqline

%matplotlib inline
```

```python
N = 1000
```

```python
x = np.linspace(-10, 10, N)
y = x + 0.6 * np.random.randn(N)
y = np.random.randn(N)
```

```python
plt.hist(y)
```

```python
f, ax = plt.subplots()

qqplot(y, ax=ax, line='s')
```
