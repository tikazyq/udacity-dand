```python
from pandas import *
import matplotlib.pyplot as plt
from ggplot import *
from pymongo import MongoClient

%matplotlib inline
```

```python
mongo = MongoClient()
db = mongo.osm

def get_df(qry):
    raw = []
    cursor = db.manchester_england.aggregate(qry)
    for doc in cursor:
        for k, v in doc.get('_id').iteritems():
            doc[k] = v
        del doc['_id']

        raw.append(doc)

    df = DataFrame(raw)

    df['month'] = to_datetime(df['timestamp'].apply(lambda x: x[:7] + '-01'))
    df['timestamp'] = to_datetime(df['timestamp'])
    df['year'] = df['timestamp'].apply(lambda x: x.year)
    df['weekday'] = df['timestamp'].apply(lambda x: x.weekday())
    
    return df
```

```python
df = get_df([
        {'$group': {'_id': {
                        'timestamp': {'$substr': ['$created.timestamp', 0, 10]}, 
                        'user': '$created.user',
                    }, 
                    'count': {'$sum': 1}}},
        {'$sort': {'_id': 1}}])

df_bing = get_df([
        {'$match': {'source': {'$regex': 'bing', '$options': 'i'}}}, 
        {'$group': {'_id': {
                        'timestamp': {'$substr': ['$created.timestamp', 0, 10]}, 
                        'user': '$created.user',
                    }, 
                    'count': {'$sum': 1}}},
        {'$sort': {'_id': 1}}])

df_yahoo = get_df([
        {'$match': {'source': {'$regex': 'yahoo', '$options': 'i'}}}, 
        {'$group': {'_id': {
                        'timestamp': {'$substr': ['$created.timestamp', 0, 10]}, 
                        'user': '$created.user',
                    }, 
                    'count': {'$sum': 1}}},
        {'$sort': {'_id': 1}}])

df_naptan = get_df([
        {'$match': {'source': {'$regex': 'naptan', '$options': 'i'}}}, 
        {'$group': {'_id': {
                        'timestamp': {'$substr': ['$created.timestamp', 0, 10]}, 
                        'user': '$created.user',
                    }, 
                    'count': {'$sum': 1}}},
        {'$sort': {'_id': 1}}])
```

```python
df.groupby('month')['count'].sum() \
    .plot(figsize=(10, 6), title='Monthly Contributions')
plt.xlabel('Time')
plt.ylabel('Contributions')
plt.show()
```

```python
df_bing.groupby('month')['count'].sum() \
    .plot(figsize=(10, 6), title='Monthly Contributions (Bing)')
plt.xlabel('Time')
plt.ylabel('Contributions')
plt.show()
```

```python
df_yahoo.groupby('month')['count'].sum() \
    .plot(figsize=(10, 6), title='Monthly Contributions (Yahoo)')
plt.xlabel('Time')
plt.ylabel('Contributions')
plt.show()
```

```python
df_naptan.groupby('month')['count'].sum() \
    .plot(figsize=(10, 6), title='Monthly Contributions (NaPTAN)')
plt.xlabel('Month')
plt.ylabel('Contributions')
plt.show()
```

```python
top_users = df.groupby('user', as_index=0)['count'].sum().sort('count', ascending=0)[:10].user.tolist()
```

```python
top_users_contributions = df[df.user.isin(top_users)] \
    .pivot_table(index='month', columns='user', values='count', aggfunc=sum).sum().order() \
    / df['count'].sum()
    
top_users_contributions.sum()
```

```python
df[df.user.isin(top_users[:]) & (df.year < 2015)].assign(_month=df.month.apply(lambda x: x.month)) \
    .groupby(['_month'], as_index=1)['count'].sum() \
    .plot(kind='bar', figsize=(10, 6), title='Contributions by Month (Top 10 Contributors)')
# plt.ylim(0, 1e5)
plt.xlabel('Month')
plt.ylabel('Contributions')
plt.show()
```

```python
df[~df.user.isin(top_users[:]) & (df.year < 2015)].assign(_month=df.month.apply(lambda x: x.month)) \
    .groupby(['_month'], as_index=1)['count'].sum() \
    .plot(kind='bar', figsize=(10, 6), title='Contributions by Month (Excluding Top 10 Contributors)')
# plt.ylim(0, 1e5)
plt.xlabel('Month')
plt.ylabel('Contributions')
plt.show()
```

```python
df[(df.year < 2015)].assign(_month=df.month.apply(lambda x: x.month)) \
    .groupby(['_month'], as_index=1)['count'].sum() \
    .plot(kind='bar', figsize=(10, 6), title='Contributions by Month')
plt.ylim(0, 2e5)
plt.xlabel('Month')
plt.ylabel('Contributions')
plt.show()
```

```python
weekdays = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

df[~df.user.isin(top_users[:]) & (df.year < 2015)] \
    .assign(weekday=df.weekday.apply(lambda x: str(x) + '. ' + weekdays[x])) \
    .groupby(['weekday'], as_index=1)['count'].sum() \
    .plot(kind='bar', figsize=(10, 6), title='Contributions by Weekday (Excluding Top 10 Contributors)')
# plt.ylim(0, 1e5)
plt.xlabel('Weekday')
plt.ylabel('Contributions')
plt.xticks()
plt.show()
```
